**Interpretation From The Box Plot**

- Every Numerical Colum has Outliers in the given data.
- Min values gives the Lowest values with respective to each columns that are present in the data.
- 25% values with respect to each column gives us the Q1 i.e quartile 1 values for that columns.
- 50% values with respect to each column gives us the Q2 i.e quartile 2 values for that columns.
- 75% values with respect to each column gives us the Q3 i.e quartile 3 values for that columns.
- Max values gives the most Max or Highest values with respective to each columns that are present in the data.
- While to calulate the Upper Extreme and Lower Extreme values we have formula
- LE = Q1-(1.5*iqr)
- UE = Q3+(1.5*iqr)
- iqr = Q3-Q1

```python
categorical_columns = df.select_dtypes(include=['object']).columns

# Create bar charts for each categorical column
for column in categorical_columns:
   
    category_counts = df[column].value_counts()                     # Create bar charts for each categorical column
    plt.figure(figsize=(10, 6))                                       # Plot bar chart
    plt.bar(x=category_counts.index, height=category_counts.values)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.xlabel(column)
    plt.ylabel('Frequency')
```

**There are outliers present in the day and brand columns in the above bar charts.**

# Standardization of Numerical Variables

**What is Standardization (Z-Score Normalization)?**

Standardization is a data preprocessing technique used to transform numerical features in a dataset so that they have a mean of 0 and a standard deviation of 1. This process ensures that all features are on a consistent scale, making them more comparable and suitable for various analytical models.

```python
standardized_data = (df[['Volume','Avg Price','Total Sales Value']] - mean) / std
```
